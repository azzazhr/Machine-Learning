# **Langkah 1 - Load Data**

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
from google.colab import drive

#Mount Google Drive
drive.mount('/content/drive')


folder_path='/content/drive/MyDrive/DATAML_AZZA/wbc.csv'
dpath=folder_path
df = pd.read_csv(dpath)
df.head()
```

# **Langkah 2 - Slice Data**

```python
df = df.drop(columns=["id", "Unnamed: 32"])
```

# **Langkah 3 - Encoding Pada Kolom Diagnosis**

```python
df["diagnosis"] = df["diagnosis"].map({"M": 1, "B": 0})
```

# **Langkah 4 - Standarisasi**

```python
scaler = StandardScaler()

# Ambil kolom numerik (selain diagnosis)
numerical_cols = df.drop(columns=["diagnosis"]).columns

# Lakukan standardisasi
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

# **Langkah 5 - Verifikasi Hasil Prepocessing**

```python
print(df.head())
```

# **Simpan Hasil Preprocessing**

```python
df.to_csv("wbc_preprocessed.csv", index=False)
```
