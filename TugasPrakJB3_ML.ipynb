**Langkah 1 - Load Data**

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
```

```python

from google.colab import drive

#Mount Google Drive
drive.mount('/content/drive')


folder_path='/content/drive/MyDrive/DATAML_AZZA/wbc.csv'
dpath=folder_path
df = pd.read_csv(dpath)
df.head()
```

**Langkah 2 - Slice Data**

```python
df = df.drop(columns=['id'])
```

**Langkah 3 - Encoding pada Kolom 'Diagnosis'**

```python
le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])

# tampilkan 5 baris pertama setelah encoding
df.head()
```

**Langkah 4 - Standarisasi Yang Memiliki Nilai Numerik**

```python
# Pisahkan fitur (X) dan target (y)
X = df.drop(columns=['diagnosis'])
y = df['diagnosis']

# Standarisasi X
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Ubah ke DataFrame biar rapi
X = pd.DataFrame(X_scaled, columns=X.columns)
```

**Langkah 5 - Stratifies Split Data**

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

#cek masing-masing ukuran data
print("Jumlah data latih:", X_train.shape[0])
print("Jumlah data uji:", X_test.shape[0])
print("\nDistribusi kelas di data latih:")
print(y_train.value_counts(normalize=True))
print("\nDistribusi kelas di data uji:")
print(y_test.value_counts(normalize=True))
```
